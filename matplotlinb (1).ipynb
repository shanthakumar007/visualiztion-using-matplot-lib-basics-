```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([2003,2004,2005,2006])
y=np.array([23,34,45,35])

plt.grid(axis="x",color="pink",linestyle="solid")

plt.title("Summa Oru Try",fontsize="14",family="Arial",color="blue",fontweight="bold")
plt.xlabel("YEAR (2023 to 2026)",fontsize="14",family="Arial",color="blue",fontweight="bold")
plt.ylabel("Values",fontsize="14",family="Arial",color="blue",fontweight="bold")
plt.xticks(x)
plt.tick_params(axis="both",labelcolor="Darkgreen")

plt.plot(x,y,marker=".",ms="20",mfc="yellow",mec="red",linestyle="dotted",linewidth="3",c="black")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

a=["fisherman","farmer","doctor","engineer"]
b=np.array([20,45,34,21])
co=["blue","green","pink","lightblue"]
plt.pie(b,labels=a,colors=co,explode=[0.1,0.1,0.1,.10],shadow="true",startangle=360,autopct="%1.1f%%")
textprops={'fontsize': 14}
textprops={"family":"Arial"}
plt.title("pie chart",fontsize="20",family="Arial",color="darkblue")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

a=np.array([0,1,2,1,3,4,5,6,7])
b=np.array([50,60,70,80,85,90,95,99,100])

a1=np.array([0,1,2,1,3,4,5,6,7])
b1=np.array([35,40,45,50,55,60,65,70,75])
plt.xlabel("Hours studied",color="blue",fontsize=20,family="Arial")
plt.ylabel("Marks oibtained",color="blue",fontsize=20,family="Arial")
plt.title("marks scored by hours studied",color="red")
plt.scatter(a,b,color="skyblue",alpha=0.5,label="class A")
plt.scatter(a1,b1,color="red",alpha=0.5,label="class B")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5])



figure,axis=plt.subplots(2,2)
axis[0,0].plot(x,x**2,color="red")
axis[0,0].set_title("round1")
plt.tick_params(axis="both",labelcolor="blue")
axis[0,1].bar(x,x**3,color="hotpink")
axis[0,1].set_title("round2")
plt.tick_params(axis="both",labelcolor="hotpink")
axis[1,0].scatter(x,x**4,color="green")
axis[1,0].set_title("round3")
plt.tick_params(axis="both",labelcolor="blue")
axis[1,1].barh(x,x**5,color="purple")
axis[1,1].set_title("round4")
plt.tight_layout()
plt.tick_params(axis="both",labelcolor="blue")
plt.show()
```
